```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv('datasets/antropometria-dataset.csv')
df.head(5)
```

```python
df.columns
```

```python
sns.boxplot(x = df['talla'],
            whis =(0,100), 
           )
```

```python
pivot=pd.pivot_table(df,
                        index=['desc_ent'],
                        values=['peso','cintura','edad'],
                        aggfunc='median',
                        columns='sexo',
                        margins=True
                       )
pivot
```

```python
sns.boxplot( x = pivot['edad', 1],
           whis = (0,100), 
           )
```

```python
sns.lineplot( data = pivot['cintura', 2]
           )
```

```python
pivot=pd.pivot_table(df,
                        index=['desc_ent'],
                        values=['ropa'],
                        aggfunc='sum',
                        columns='sexo',
                        margins=True
                       )
pivot.head(10)
```

```python
pivot10 =pivot.sort_values(by =[( 'ropa',     2)],
                           ascending = False, #
                           axis = 0)
pivot10.head(10)
```

```python
pivot10=pivot10.head(11)
pivot10=pivot10.drop(['All'],
                             axis= 0)
```
