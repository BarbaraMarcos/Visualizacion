```python
df = pd.read_csv('datasets/nhanes_2015_2016.csv')
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import statsmodels.api as sm
import numpy as np
```

```python
df.columns
```

```python
vars = ['SEQN', 'RIAGENDR', 'RIDAGEYR', 'BMXWT', 'BMXHT', 'BMXBMI', 
       'BMXWAIST', 'HIQ210']
```

```python
df = df[vars].dropna()
df
```

```python
df.corr()
```

### BMXWT = b0 + b1*RIDAGEYR

```python
model = sm.OLS.from_formula('BMXWT~RIDAGEYR', data = df)
result = model.fit()
result.summary()
```

```python
df['RiaGENDR'] = df.RIAGENDR.replace({1:'Male',2:'Female'})
df
```

### BMXWT = b0 + b1*RIDAGEYR + b2*RiaGENDR
### BMXWT = b0 + b1*RIDAGEYR + b2(Male + Female)
### Female
### BMXWT = 77.5710 - 0.0228*RIDAGEYR + 10.5351(Male)

```python
model = sm.OLS.from_formula('BMXWT~RIDAGEYR + RiaGENDR', data = df)
result = model.fit()
result.summary()
```
