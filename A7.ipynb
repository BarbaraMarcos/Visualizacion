```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
%ls datasets/
```

```python
df = pd.read_csv('datasets/antropometria-dataset.csv')
df.head()
```

```python
df.columns
```

```python
peso = df['sistol'].plot(kind = 'hist', bins = 15)
```

```python
sns.lmplot(x='diastol',
           y= 'peso',
           hue = 'sexo',
           data = df,
           fit_reg = False, 
          )
```

```python
sns.boxplot(data = df.loc[df['sexo'] == 1,
                             ['cintura']],
           ).set_title('Box plot de Género 1 (M)')
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(1, 2, 1) #1 columna, 2 row, posición 1
sns.boxplot(data = df.loc[df['sexo'] == 1, ['cintura']],
           ).set_title('Box plot sexo 1')

ax2 = fig.add_subplot(1, 2, 2) #1 columna, 2 row, posición 2
sns.boxplot(data = df.loc[df['sexo'] == 2, ['cintura']],
           ).set_title('Box plot sexo 2')
```

```python
sns.lmplot(data = df.loc[df['sexo'] == 2],
           x='sistol',
           y= 'diastol',
           hue = 'sexo',
           fit_reg = False, 
          )
```

```python
sns.histplot(data = df.loc[df['sexo'] == 2,
                             ['cintura']],
             bins = 20,
           )
```
