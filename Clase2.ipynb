## loc y iloc
[row iicial: row final, col inicial: col final]
si usa loc, usa nombres de los renglones y nombres de las columnas.
si usa iloc, usa ínfices de los renglones y las columnas

```python
df.iloc[2:10,3:6]
```

```python
df.loc[2:7,'Age':'Height']
```
