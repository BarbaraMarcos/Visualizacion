```python
df = pd.read_excel('datasets/Canada.xlsx',
                   sheet_name = 'Canada by Citizenship',
                   engine = 'openpyxl',
                   skiprows = range(20),
                   skipfooter = 2
                  )
df
```

```python
df.columns
```

```python
all(isinstance(column, str) for column in df.columns)
```

```python
df.columns = list(map(str, df.columns))
all(isinstance(column, str) for column in df.columns)
```

```python
df.drop(['Type',    'Coverage', 'REG',        'DEV',
           'DevName', 'Unnamed: 43',
       'Unnamed: 44', 'Unnamed: 45', 'Unnamed: 46', 'Unnamed: 47',
       'Unnamed: 48', 'Unnamed: 49', 'Unnamed: 50'], inplace = True,
        axis = 1)
df.columns
```

```python
df.rename(columns = {'OdName': 'Country', 
                     'AreaName': 'Continent', 
                     'RegName': 'Region',
                    },
          inplace = True)
df.columns
```

```python
df['Total'] = df.iloc[:, 4:].sum(axis = 1)
df.head()
```

```python
df.set_index('Country', inplace = True)
df.head()
```

```python
df.loc['Japan', 'Total']
```

```python
years = list(map(str, range(1980,2014)))
years
```

```python
df.loc['Japan', years].sum()
```

## Me gustaría saber cuántas personas llegaron desde Haití a Canadá desde 1980 hasta el 2013, con una gráfica de lineas

```python
df.loc['Haiti', years].plot()
plt.xlabel('X', fontsize = 16)
plt.xticks(fontsize = 16, rotation = 45)

plt.ylabel('Número de inmigrantes', fontsize = 16)
plt.yticks(fontsize = 16, rotation = 45)

plt.title('Número de inmigrantes desde Haití a Canadá de 1980 - 2013', fontsize = 17)
plt.text(30, 6500, 'Terremoto', fontsize = 14)
plt.show()
```

## Conclusión
La inmigración desde 1980 hasta 2010 fue en un rango desde 1500 -3500, pero en 2009 hubo un incremento
